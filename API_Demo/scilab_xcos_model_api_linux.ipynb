```python
import math
import json
import pandas as pd
import subprocess
```

```python
REQUEST = json.dumps({
'path' : {},
'args' : {}
})
```

```python
# GET /scilab_xcos_model
req = json.loads(REQUEST)
args = req['args']

# make sure there is an argument
if 'Amp' in args:
    Amp = str(args['Amp'][0])
else: 
    Amp = 3
    
if 'Freq' in args:
    Freq = str(args['Freq'][0])
else:
    Freq = 1
        
# Execute the xcos model and fetch string array of output
execString = (["/tmp/scilab-6.0.2/bin/scilab-adv-cli", "-args", str(Amp), str(Freq), "-f", "/root/jupyter_demos/API_Demo/models/scilab_sine2.sce", "-nb", "-nw", "-quit"])
#print(execString)
result = subprocess.run(execString, capture_output=True, text=True)
print(result.stdout)
```
