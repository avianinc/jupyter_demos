```python
# gets the list of users from the db

import os
import pymysql
import pandas as pd

host = os.getenv('MYSQL_HOST') # or localhost
port = os.getenv('MYSQL_PORT') # typically 3306
user = os.getenv('MYSQL_USER') # root
password = os.getenv('MYSQL_PASSWORD') # 123456
database = os.getenv('MYSQL_DATABASE') # mysql

conn = pymysql.connect(
    host="localhost",
    port=int(3306),
    user="root",
    passwd="123456",
    db="mysql",
    charset='utf8mb4')

df = pd.read_sql_query("SELECT * FROM mysql.user;", conn)
df.tail(10)
```
